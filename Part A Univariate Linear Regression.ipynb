```python
import numpy as np
import pandas as pd
```

```python
file_train = 'cancer_us_county-training.csv'
file_test = 'cancer_us_county-testing.csv'

df_train = pd.read_csv(file_train)
df_test = pd.read_csv(file_test)
```

## Hypothesis 1:  

Cancer Mortality rate is higher for those counties in the United States that has lower Median Income.

```python
df_cleaned_train = df_train.copy()
df_cleaned_test = df_test.copy()
```

```python
df_cleaned_train.plot(x = 'medIncome', y = 'TARGET_deathRate', kind = 'scatter')
```

```python
y_train = df_cleaned_train['TARGET_deathRate'].values
y_test = df_cleaned_test['TARGET_deathRate'].values

X_train = df_cleaned_train['medIncome'].values
X_test = df_cleaned_test['medIncome'].values
```

```python
y_mean = y_train.mean()
y_base = np.full(y_train.shape, y_mean)
```

```python
(y_base - y_train).sum()
```

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
```

```python
print(mse(y_train, y_base))
print(mae(y_train,y_base))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
X_train.reshape(-1,1)
```

```python
y_train
```

```python
reg.fit(X_train.reshape(-1,1), y_train)
```

```python
y_preds_train = reg.predict(X_train.reshape(-1,1))
```

```python
print(mse(y_train, y_preds_train, squared=False))
print(mae(y_train, y_preds_train))
```

```python
y_preds_test = reg.predict(X_test.reshape(-1,1))
```

```python
print(mse(y_test, y_preds_test, squared=False))
print(mae(y_test, y_preds_test))
```

```python
import altair as alt

perfect = alt.Chart(pd.DataFrame({'target': y_train, 'preds': y_train})).mark_line(color='green').encode(
    x='target',
    y='preds'
)

pred_chart = alt.Chart(pd.DataFrame({'target': y_train, 'preds': y_preds_train})).mark_line().encode(
    x='target',
    y='preds'
  )

pred_chart + perfect
```

```python
perfect = alt.Chart(pd.DataFrame({'target': y_test, 'preds': y_test})).mark_line(color='green').encode(
    x='target',
    y='preds'
)

pred_chart = alt.Chart(pd.DataFrame({'target': y_test, 'preds': y_preds_test})).mark_line().encode(
    x='target',
    y='preds'
  )

pred_chart + perfect
```
